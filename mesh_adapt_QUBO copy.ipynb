```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define grid range and resolution
x = np.linspace(-1, 1, 100)  # 50 points from -5 to 5 in x
y = np.linspace(-1, 1, 100)  # 50 points from -5 to 5 in y

# Generate meshgrid
M = np.meshgrid(x, y)
X = M[0]
Y = M[1]

# Define the function f(x, y) = sin(sqrt(x^2 + y^2))
Z = np.sin(2*np.pi*X)

# Create a 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, M[1], Z, cmap='viridis', edgecolor='k')

# Labels and title
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis (f(x, y))")
ax.set_title("3D Surface Plot of f(x, y)")

plt.show()
print(M[0])
```
